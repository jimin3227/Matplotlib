# 스타일

- align: 눈금과 막대의 위치
- edgecolor: 막대 테두리 색
- linewidth: 테두리의 두께
- tick_label: 틱에 문자열 지정 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.bar(x, values, align='edge', edgecolor='lightgray',
        linewidth=5, tick_label=years)

plt.show()
```
