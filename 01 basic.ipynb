# matplotlib.pyplot

## y만 입력

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4]) # y값, x값은 [0, 1, 2, 3] 자동 생성
plt.show()
```

## x, y 입력

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16]) # x값, y값
plt.show()
```

# 여러개 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0., 5., 0.2) # 0, 0.2, 0.4 ..

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
