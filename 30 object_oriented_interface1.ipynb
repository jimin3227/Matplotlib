# plt.subplots()

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
plt.show()
```

# 행, 열

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 2)
plt.show()
```

# 축 공유

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)
plt.show()
```
