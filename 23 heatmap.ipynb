# 기본

```python
import matplotlib.pyplot as plt
import numpy as np

arr = np.random.standard_normal((30, 40))

plt.matshow(arr)

plt.show()
```

# 컬러바 

colorbar
- shrink: 컬러바의 크기
- asspect: 컬러바의 종횡비(default: 20)

```python
import matplotlib.pyplot as plt
import numpy as np

arr = np.random.standard_normal((30, 40))

plt.matshow(arr)
plt.colorbar()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

arr = np.random.standard_normal((30, 40))

plt.matshow(arr)
plt.colorbar(shrink=0.6, aspect=30)

plt.show()
```

# 범위 

```python
import matplotlib.pyplot as plt
import numpy as np

arr = np.random.standard_normal((30, 40))

plt.matshow(arr)
plt.colorbar(shrink=0.8, aspect=10)
plt.clim(-1.0, 1.0) # -1보다 작거나 1보다 큰 값에 대해서는 -1,1과 같은 색

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

arr = np.random.standard_normal((30, 40))

plt.matshow(arr)
plt.colorbar(shrink=0.8, aspect=10)
plt.clim(-3.0, 3.0) # -3보다 작거나 3보다 큰 값에 대해서는 -3,3과 같은 색

plt.show()
```

# 색상

```python
import matplotlib.pyplot as plt
import numpy as np

arr = np.random.standard_normal((30, 40))
cmap = plt.get_cmap('bwr')

plt.matshow(arr, cmap=cmap)
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

arr = np.random.standard_normal((30, 40))
cmap = plt.get_cmap('Greys')

plt.matshow(arr, cmap=cmap)
plt.colorbar()
plt.show()
```
