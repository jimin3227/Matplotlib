# 기본

```python
import matplotlib.pyplot as plt

years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.barh(years, values)

plt.show()
```

# 색상

```python
import matplotlib.pyplot as plt

years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.barh(years, values, color='y')

plt.show()
```

```python
import matplotlib.pyplot as plt

years = ['2018', '2019', '2020']
values = [100, 400, 900]
colors = ['y', 'dodgerblue', 'C2']

plt.barh(years, values, color=colors)

plt.show()
```

# 폭

```python
import matplotlib.pyplot as plt

years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.barh(years, values, height=0.4)

plt.show()
```

# 스타일

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.arange(3)
years = ['2018', '2019', '2020']
values = [100, 400, 900]

plt.barh(y, values, align='edge', edgecolor='#eee',
         linewidth=5, tick_label=years)

plt.show()
```
