# 범위

- plt.xlim([x_min, x_max]): X축이 표시되는 범위를 지정
- plt.ylim([y_min, y_max]): Y축이 표시되는 범위를 지정
- plt.axis([x_min, x_max, y_min, y_max]): X, Y축이 표시되는 범위를 지정

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xlim([0, 5])     
plt.ylim([0, 20])   

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.axis([0, 5, 0, 20])

plt.show()
```

# 범위 얻기

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

x_range, y_range = plt.xlim(), plt.ylim()
print(x_range, y_range)

axis_range = plt.axis()
print(axis_range)

plt.show()
```
