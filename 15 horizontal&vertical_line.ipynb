# 수평선 

- plt.axhline(y, xmin, xmax): xmin, xmax 0(좌측 끝)~1(우측 끝)
- plt.hlines(y, xmin, xmax)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=2)
plt.hlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3)

plt.show()
```

# 수직선

- plt.axvline(x, ymin, ymax): ymin, ymax 0(하단 끝)~1(상단 끝)
- plt.hlines(x, ymin, ymax)

```python
import matplotlib.pyplot as plt

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axvline(1.0, 0.2, 0.8, color='lightgray', linestyle='--', linewidth=2)
plt.vlines(1.8, -3.0, 2.0, color='gray', linestyle='solid', linewidth=3)

plt.show()
```
